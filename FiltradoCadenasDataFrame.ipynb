```python
import re
from unidecode import unidecode
# unidecode() se usa para eliminar acentos y caracteres especiales. Convierte caracteres como 
# "á", "é", "í", "ó", "ú" a "a", "e", "i", "o", "u". 
# Es útil para normalizar texto y hacer comparaciones que ignoren acentos y caracteres especiales.

# Función para normalizar el texto
def normalizar_cadena(texto, cambio = str):
    # pd.isna() verifica si un valor es NaN (Not a Number), None, o algún otro tipo de valor nulo.
    if pd.isna(texto):
        return texto
    
    # Convertir a string por si hay números u otros tipos
    texto = str(texto)
    
    # Normalizar el texto: quitar acentos, convertir a minúsculas y eliminar espacios extras
    texto_normalizado = unidecode(texto).lower().strip()
    texto_normalizado = re.sub(r'\s+', '', texto_normalizado)
    # replace(cambiar_por_x,x) se usa para remplazar caracteres específicos.
    texto_normalizado = texto_normalizado.replace(" ", "").replace("-", "").replace("_", "").replace(".", "").replace(",", "")
    
    # Verificar si el texto normalizado coincide con "cambio"
    if texto_normalizado == unidecode(cambio.replace(" ", "")).lower():
        return str(cambio)
    else:
        return texto

# Mostrar todo el dataframe sin truncar (visualización)
pd.set_option('display.max_rows', None)  # Mostrar todas las filas
pd.set_option('display.max_columns', None)  # Mostrar todas las columna


# Aplicar la función a la columna del DataFrame
dtf["estu_prgm_academico"] = dtf['estu_prgm_academico'].apply(normalizar_cadena, cambio="CONTADURÍA PÚBLICA").to_frame()

# Mostrar los primeros 1000 resultados del DataFrame modificado
#dtf.head(1000)

# Función para cambiar una palabra por otra en una columna del DataFrame
def cambiar_palabra_por_otra(texto, palabra_vieja, palabra_nueva):
    if pd.isna(texto):
        return texto
    texto = str(texto)
    texto = texto.replace(palabra_vieja, palabra_nueva)
    return texto

# Mostrar todo el dataframe (sin truncar)
pd.set_option('display.max_rows', None)  # Mostrar todas las filas
pd.set_option('display.max_columns', None)  # Mostrar todas las columna        
# Aplicar la función a la columna del DataFrame y reemplazar las palabras                             
dtf["estu_prgm_academico"] = dtf["estu_prgm_academico"].apply(cambiar_palabra_por_otra, palabra_vieja="INGENIERIA", palabra_nueva="INGENIERÍA").to_frame()
dtf["estu_prgm_academico"] = dtf["estu_prgm_academico"].apply(cambiar_palabra_por_otra, palabra_vieja="INGENIERÍA\x8d", palabra_nueva="INGENIERÍA").to_frame()


# Se ha encontrado un valor atípico en específico en la columna "estu_nucleo_pregrado", para seleccionarlo y cambiarlo:
#dtf.loc[0, "estu_nucleo_pregrado"] el cual es 'INGENIERÃ\x8dA INDUSTRIAL Y AFINES'
dtf["estu_nucleo_pregrado"] = dtf["estu_nucleo_pregrado"].apply(cambiar_palabra_por_otra, palabra_vieja="INGENIERÃ\x8dA", palabra_nueva="INGENIERÍA").to_frame()

#dtf.head(1000)



# Función para normalizar varias textos de una columna del DataFrame
def normalizar_varias_cadenas(df, columna, columna1 = None, columna2 = None, columna3 = None, columna4 = None, columna5 = None, columna6 = None, cambio = str):
    df[columna] = df[columna].apply(normalizar_cadena, cambio=cambio).to_frame()
    if columna1:
        df[columna1] = df[columna1].apply(normalizar_cadena, cambio=cambio).to_frame()
    elif columna2:
        df[columna2] = df[columna2].apply(normalizar_cadena, cambio=cambio).to_frame()
    elif columna3:
        df[columna3] = df[columna3].apply(normalizar_cadena, cambio=cambio).to_frame()
    elif columna4:
        df[columna4] = df[columna4].apply(normalizar_cadena, cambio=cambio).to_frame()
    elif columna5:
        df[columna5] = df[columna5].apply(normalizar_cadena, cambio=cambio).to_frame()
    elif columna6:
        df[columna6] = df[columna6].apply(normalizar_cadena, cambio=cambio).to_frame()
    else:
        None
    return df

# Función para cambiar una palabra por otra en varias columnas del DataFrame
def cambiar_varias_palabras(df, columna1, columna2=None,columna3=None, columna4=None, columna5=None, columna6=None, palabra_vieja1 = str, palabra_nueva1 = str):
    df[columna1] = df[columna1].apply(cambiar_palabra_por_otra, palabra_vieja=palabra_vieja1, palabra_nueva=palabra_nueva1).to_frame()
    if columna2:
        df[columna2] = df[columna2].apply(cambiar_palabra_por_otra, palabra_vieja=palabra_vieja1, palabra_nueva=palabra_nueva1).to_frame()
    elif columna3:
        df[columna3] = df[columna3].apply(cambiar_palabra_por_otra, palabra_vieja=palabra_vieja1, palabra_nueva=palabra_nueva1).to_frame()
    elif columna4:
        df[columna4] = df[columna4].apply(cambiar_palabra_por_otra, palabra_vieja=palabra_vieja1, palabra_nueva=palabra_nueva1).to_frame()
    elif columna5:
        df[columna5] = df[columna5].apply(cambiar_palabra_por_otra, palabra_vieja=palabra_vieja1, palabra_nueva=palabra_nueva1).to_frame()
    elif columna6:
        df[columna6] = df[columna6].apply(cambiar_palabra_por_otra, palabra_vieja=palabra_vieja1, palabra_nueva=palabra_nueva1).to_frame()

    return df    

# Ejemplo de uso:
dtf = normalizar_varias_cadenas(dtf, 'estu_prgm_academico', 'estu_nucleo_pregrado', cambio="CONTADURÍA PÚBLICA")
dtf = normalizar_varias_cadenas(dtf, 'estu_depto_presentacion', 'estu_inst_departamento', "estu_inst_municipio", "estu_mcpio_presentacion", "estu_prgm_departamento", "estu_prgm_municipio",  cambio="BOGOTÁ")

""" estu_depto_presentacion  estu_inst_departamento  estu_inst_municipio	estu_mcpio_presentacion estu_prgm_municipio 'BOGOTÃ\x81'"""

# Ejemplo de uso:
dtf = cambiar_varias_palabras(dtf, 'estu_inst_municipio', "estu_mcpio_presentacion", "estu_prgm_municipio", "estu_prgm_municipio", palabra_vieja1='BOGOTÃ\x81 D.C.', palabra_nueva1="BOGOTÁ D.C.")
dtf = cambiar_varias_palabras(dtf, 'estu_prgm_departamento', palabra_vieja1='BOGOTÃ\x81', palabra_nueva1="BOGOTÁ")

# Visualizar los primeros 100 registros
dtf.head(49)


```
